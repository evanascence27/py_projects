```python
import seaborn as sn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame(np.random.random((6,7)), columns = ['a', 'b', 'c', 'd', 'e', 'f', 'g']);
#custom heatmap
sn.heatmap(df, cmap = "YlGn", annot=True, annot_kws={'size':7}, linewidths=2.0,linecolor='gray', cbar=False)
plt.show()
```

```python
#making custom heatmap to demonstrate normalization
df = pd.DataFrame(np.random.randn(6,7) * 6 + 37);
df[1] = df[1] + 60;
sn.heatmap(df, cmap = "plasma");
plt.show();
```

```python
#normalizing
df_norm = (df - df.mean()) / df.std();
sn.heatmap(df_norm, cmap = "plasma");
plt.show()
```
