```python
#detecting edges in image using Canny()
cv2.imwrite("edges_j_vermeer.jpg", cv2.Canny(color_img,602,315))
cv2.imshow("Edges", cv2.imread("edges_j_vermeer.jpg"))
key = cv2.waitKey(0) & 0xFF
if key == 27:
    cv2.destroyAllWindows()
elif key == ord('s'): 
    cv2.imwrite("F:/AIProject/myProjects/images/vermeer_colored.jpg", color_img)
    cv2.destroyAllWindows()
```

```python
#detecting faces in CV using haarcascade_frontalface_default.xml on our machine
img = cv2.imread("ellen.jpg");
cv2.imwrite("edges_ellen.jpg", cv2.Canny(img, 602, 315));
cv2.imshow("face detection edges", cv2.imread("edges_ellen.jpg"))
cv2.waitKey(0);
fd = cv2.CascadeClassifier("C:\\Users\\KIIT\\Anaconda3\\Lib\\site-packages\\cv2\\data\\haarcascade_frontalface_default.xml");
img2gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY);
#detection
faces = fd.detectMultiScale(img2gray, 1.5, 4);
for (x, y, w, h) in faces:
    img = cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 3);
cv2.imwrite("face_ellen.jpg", img);
#show detected face
cv2.imshow("face detection", img);
key = cv2.waitKey(0) & 0xFF
if key == 27:
    cv2.destroyAllWindows()
elif key == ord('s'): 
    cv2.imwrite("F:/AIProject/myProjects/images/vermeer_colored.jpg", color_img)
    cv2.destroyAllWindows()
```

```python
#eye detection
eye_cascade=cv2.CascadeClassifier("C:\\Users\\KIIT\\Anaconda3\\Lib\\site-packages\\cv2\\data\\haarcascade_eye.xml")
eyes=eye_cascade.detectMultiScale(img2gray, 1.002 , 1);
for (ex,ey,ew,eh) in eyes:
    img = cv2.rectangle(img, (ex,ey), (ex+ew,ey+eh), (255,0,0), 2)
#show detected eyes
cv2.imshow("face detection", img);
key = cv2.waitKey(0) & 0xFF
if key == 27:
    cv2.destroyAllWindows()
elif key == ord('s'): 
    cv2.imwrite("F:/AIProject/myProjects/images/vermeer_colored.jpg", color_img)
    cv2.destroyAllWindows()
```
